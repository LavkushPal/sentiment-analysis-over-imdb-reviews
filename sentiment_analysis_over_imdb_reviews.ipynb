```python
from keras.datasets import imdb
from keras.utils import pad_sequences
from keras.models import Sequential
from keras.layers import Dense,SimpleRNN,Embedding,Flatten,Input
```

```python
(X_train,y_train),(X_test,y_test)=imdb.load_data(num_words=10000)
```

```python
X_train.shape
```

```python
X_train=pad_sequences(X_train,maxlen=500,padding='post')
X_test=pad_sequences(X_test,maxlen=500,padding='post')
```

```python
X_train
```

```python
model = Sequential()

model.add(Input(shape=(500,)))
model.add(Embedding(10000, 64))

model.add(SimpleRNN(32, activation='tanh'))

model.add(Dense(1, activation='sigmoid'))
```

```python
model.summary()
```

```python
model.compile(optimizer='adam', loss='binary_crossentropy', metrics=['accuracy'])
```

```python
model.fit(X_train, y_train, epochs=5, validation_data=(X_test, y_test))
```

```python
from tensorflow.keras.preprocessing.text import Tokenizer
```

```python
#find integer encoding and do the padding
#it is ready to predict now

tokenizer = Tokenizer(oov_token='<nothing>')

docs=["this movie was hilarious and too funny"]
tokenizer.fit_on_texts(docs)

sequences = tokenizer.texts_to_sequences(docs)
sequences = pad_sequences(sequences,padding='post',maxlen=500)

#print(sequences)

model.predict(sequences)
```
